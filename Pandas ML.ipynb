```python
#Handling Missing Data
```

```python
df = pd.DataFrame({'A':[1,2,np.nan,4],
                  'B':[5,np.nan,np.nan,8],
                  'C':[10,20,30,40]})
```

```python
df
```

```python
df.dropna()
```

```python
df.dropna(axis=1)
```

```python
df
```

```python
df.dropna(thresh= 3,axis=1)
```

```python
df.fillna(value='fill')
```

```python
df['A'].fillna(df['A'].mean())
```

```python
df = df.fillna(df.mean())
```
