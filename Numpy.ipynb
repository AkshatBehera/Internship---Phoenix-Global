```python
#Maps and Filters
#Map: Takes everything that is passed
#Filter : Gives results based on the condition passed
```

```python
#Map function

num_var = [1,2,3,4,5]
def pow_func(par):
  return pow(par,par-1)

def my_map_func(the_func, the_iterable):
  res=[]
  for i in the_iterable:
    res.append(the_func(i))
  return res
```

```python
map_result1 = my_map_func(pow_func,num_var)
print("My map function result is:{}".format(map_result1))

map_result2 = list(map(lambda x : pow(x,x-1),num_var))
print('My In-built function result is:{}'.format(map_result2))
```

```python
#Filter 
def cond_func(par):
  return par%2

def my_filter_func(the_func,the_iterable):
  res= []
  for i in the_iterable:
    if the_func(i)==0:
      res.append(i)
    else:
      pass
  return tuple(res)
```

```python
filter_result1 = my_filter_func(cond_func,num_var)
print('My filter function result is : {}'.format(filter_result1))
filter_result2 = tuple(map(lambda x: x%2==0,num_var))
print('My inbuilt function result is: {}'.format(filter_result2))
```
